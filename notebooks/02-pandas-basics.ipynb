```python
data = {
        'Name' : ['Randy', 'Dandy', 'Handy', 'Pandy', 'Sandy', 'Sundy'],
        'Grade' : [60, 70, 80, 90, 90, 85],
        'Subject' : ['Maths', 'Science', 'English', 'Psychology', 'IT', 'Maths'],
        'Passed' : [False, False, True, True, True, True]
        }

df = pd.DataFrame(data)
print(df)
print("\n Shape :", df.shape)
print("Column Names:", df.columns.tolist())
print("Data Types: ", df.dtypes)
print("First Four Rows :", df.head(4))
print("Data Frame Info :", df.info())
```

```python
# Selecting a single column
print("Names only:")
print(df['Name'])

# Selecting multiple columns
print("\nName and Grade only:")
print(df[['Name', 'Grade']])

# Filtering rows based on condition
print("\nStudents who passed:")
print(df[df['Passed'] == True])

# Filtering with a number condition
print("\nStudents with grade above 80:")
print(df[df['Grade'] > 80])

# Multiple conditions
print("\nStudents who passed AND grade above 80:")
print(df[(df['Passed'] == True) & (df['Grade'] > 80)])
```

```python
import numpy as np

data = {
    'Name':   ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age':    [25, None, 35, 28, None],
    'Salary': [50000, 60000, None, 55000, 45000],
    'City':   ['New York', 'Boston', None, 'Seattle', 'Austin']
}

df = pd.DataFrame(data)
print(df)
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Option 1 - Drop rows with any missing values
df_dropped = df.dropna()
print("After dropping missing rows:")
print(df_dropped)

# Option 2 - Fill missing values with a fixed value
df_filled = df.fillna("Unknown")
print("\nAfter filling with Unknown:")
print(df_filled)

# Option 3 - Fill missing values with the mean (most common in ML)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nAfter filling numbers with mean:")
print(df)

# Verify no more missing values
print("\nMissing values now:")
print(df.isnull().sum())
```

```python
import pandas as pd
data = {
    'Product':  ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price':    [999, None, 499, None, 79],
    'Stock':    [10, 5, None, 8, None],
    'Category': ['Electronics', None, 'Electronics', 'Electronics', None]
}

df = pd.DataFrame(data)
print(df)
print(df.isnull().sum())
df['Price'] = df['Price'].fillna(df['Price'].mean())
df['Stock'] = df['Stock'].fillna(0)
df['Category'] = df['Category'].fillna('Unknown')
print(df.isnull().sum())

data = {
    'Name' : ['Ron', 'Von', None, 'San'],
    'Age' : [56, None, 89, 34],
    'Gender' : ['Male', 'Male', 'Female', 'Female'],
    'Grade' : [1, 4, 5, 1]

}
df = pd.DataFrame(data)
print(df.shape)
drop_df = df.dropna()
print(drop_df)
print(drop_df.shape)
```
